```python
import itertools
import random
import time

probability_package_recive = 60
```

definindo a classe Package

```python
class Package():

    newid = itertools.count().__next__
    
    def __init__(self): 
        self.id = Package.newid()
        self.send = False
        self.recivedACK = False
        
```

definindo a classe ACK 

```python
 class ACK():
        
    def __init__(self, idPackage):
        self.idPackage = idPackage.id
        self.response  = False
        self.recivePackage = False
```

```python
def receberACK(package, ack):
    
    alpha = random.randrange(1,100)
    print("ACK do pacote : " + str(ack.idPackage))
    print("Pacote recebido no servidor!")
    print("Enviando o ACK para o cliente...")
    global number_of_packages_sent
    ack.response = True
    if alpha < probability_package_recive:
        package.recivedACK = True
        isPackageSent = (package.send is True)
        isPackageRecived = (ack.recivePackage is True)
        isACKSent = (ack.response is True)
        isACKRecived = (package.recivedACK is True)
        if(isPackageRecived and isPackageSent and isACKRecived and isACKSent):
            print("ACK Recebido. Pacote enviado com sucesso!")
        else:
            print("erro")
            print("isPackageRecived", "isPackageSent", "isACKRecived", "isACKSent")
            print(isPackageRecived, isPackageSent, isACKRecived, isACKSent)
    else:
        print("Aguardando pacote...")
        print("Pacote não enviado. Reenviando...")
        package.send = False
        ack.recivePackage = False
        enviarPackage(package, ack)
        
```
