Definindo os imports e constantes

```python
import itertools
import random
from time import sleep
probability_package_recive = 90
number_of_packages = 100
```

    Definindo a classe Package

```python
class Package: 
    
    newid = itertools.count().__next__
    
    def __init__(self):
                        
        self.id = Package.newid()
        self.sentPackage = False
        self.recivedACK = False
```

    Definindo a classe ACK

```python
class ACK: 
    
    def __init__(self, package):
        
        self.idPackage = package.id
        self.recivedPackage = False
        self.idPackagesentACK = False
        
```

definindo o envio do pacote

```python
def enviarPackage(package, ack):
    alpha = random.randrange(1,100)
    print("Pacote = " + str(package.id))
    print("Enviando pacote...")
    package.sentPackage = True
    if alpha < probability_package_recive:
        ack.recivedPackage = True
        print("Pacote enviado pelo cliente!")
```

definindo a recepção do ACK

```python
def receberACK(package, ack):

    alpha = random.randrange(1,100)
    print("ACK do pacote : " + str(ack.idPackage))
    print("Enviando o ACK para o cliente...")
    ack.sentACK = True
    if alpha < probability_package_recive:
        package.recivedACK = True

    
```

Gerando pacotes

```python
def generatePackages(number_of_packages):
    list_packages = []
    list_ACK = []
    
    for pack in range(number_of_packages):
        
        package = Package()
        ack = ACK(package)
        list_packages.append(package)
        list_ACK.append(ack)
    
    return[list_packages, list_ACK]
```
